```python
import numpy as np

# Create array using Python list
arr = np.array([10, 20, 30, 40, 50])

# Display array
print("Array:", arr)

# Shape of array
print("Shape:", arr.shape)

# Data type
print("Data type:", arr.dtype)
```

```python
import numpy as np

# 3x4 matrix of zeros
zeros_matrix = np.zeros((3,4))

# 2x5 matrix of ones
ones_matrix = np.ones((2,5))

print("Zeros Matrix:")
print(zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\nOnes Matrix:")
print(ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python
import numpy as np

# Create array from 1 to 12
arr = np.arange(1,13)

print("Original Array:")
print(arr)
print("Shape:", arr.shape)

# Reshape to 3x4
reshaped = arr.reshape(3,4)

print("\nReshaped Array:")
print(reshaped)
print("Shape:", reshaped.shape)
```

```python
import numpy as np

# Random integers
arr = np.random.randint(1,100,10)

print("Array:", arr)

print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
```
